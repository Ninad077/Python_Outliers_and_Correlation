# Outliers and Correlation

Outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("advertising.csv")
```

```python
data
```

```python
# Let us plot a boxplot to understand the concept of outliers.
# Outliers are the values that skews away from the original dataset. In simple words, the value that are 
# deviation from original dataset.

# Let us first start plotting boxplot for all the columns in data.

column = data["TV"]
plt.boxplot(column)
```

```python
# Let us understand the above o/p (boxplot) now.
# So in the above boxplot, the upper and lower horizontal lines are called upper & lower whiskers.
# The line inside the box is the median and the box itself indicates how many values lie close to the median.

# For the above diagram, we've no outliers. Let us check another column
```

```python

column = data["Radio"]
plt.boxplot(column)
```

```python
# Again a decent boxplot without any outliers.
# Now let us check the column 'Newspaper'

column = data["Newspaper"]
plt.boxplot(column)
```

```python
# As evident from the figure above, there are 2 co-ordinates that skew/deviate from the original data. These co-ordinates
# are called as Outliers.
# Any value that lies beyound the upper and lower whiskers are called as Outliers.

# So now, if we want to eliminate the outlier in this case, we've to consider the data that lies below the threshold of
# the upper whisker (which is 90 in this case)

data_new = data[data["Newspaper"]< 90]
```

```python
data_new
```

```python
plt.boxplot(data_new["Newspaper"])
```

Correlation

```python
# Correlation tells how different columns in the dataset are related to each other.
# Correlation defn : One unit change in one coulmn will translate to how many units change in another column.
# 3 types of Correlation:
    # Positive correlation: Direct proportionality b/w 2 columns. 
                            # If one increases another increases, if one decreases another decreases.
    # Negative correlation: Inverse proportionality b/w 2 columns. 
                            # If one increases another decreases, if one decreases another increases.
    # No correlation      : No relation or dependancy b/w 2 columns. 
                            # One's increase or decrease does not affect the other column.
```

```python
# Positive Correlation

# To check for an example of positive correlation, let us see the following example.

column_a = data["TV"]
column_b = data["Sales"]

plt.scatter(column_a,column_b)
plt.xlabel("TV")
plt.xlabel("Sales")
plt.title("TV vs Sales: Positive Correlation")
```

```python
# Here as evident from the o/p above there exists are a strong positive correlation b/w TV and Sales.
```

```python
# Let us check the same for the column "Radio" & "Newspaper"

column_c = data["Radio"]
column_d = data["Sales"]

plt.scatter(column_c,column_d)
plt.xlabel("Radio")
plt.xlabel("Sales")
plt.title("Radio vs Sales")
```

```python
column_e = data["Newspaper"]
column_f = data["Sales"]

plt.scatter(column_e,column_f)
plt.xlabel("Newspaper")
plt.xlabel("Sales")
plt.title("Newspaper vs Sales")
```

```python
# As seen for both the the columns "Newspaper" & "Radio" there does not exist a strong correlation 
# b/w these columns wrt Sales.
```

```python
# The value of Correlation spans b/w -1 0 and 1
# -1 or closer values: Negative correlation
#  0 or closer values: No correlation
#  1 or closer values: Positive correlation

# But to get an exact match of correlation we've Heat Map

import seaborn as sns
sns.heatmap(data.corr(), annot=True) # annot means annotation, it will assign values to the heatmap.
```

```python
# Let us interpret the image above.
# The strip on it's right side gives the value of correlation.
# Lighter shade (values closer to 1.0) gives higher correlation and Darker shade (values closer to 0.2) gives 
# lower correlation
# The data in the diagonal is howing 1 always because it shows the correlation of the column with itself.
# For e.g Sales (X-axis) = Sales (Y- axis), Newspaper (X-axis) = Newspaper (Y- axis)... and so on....

# If we check the correlation of different coulns wrt Sales we get
    # 0.16 for Newspaper
    # 0.35 for Radio
    # 0.9 for TV (close to 1 hence a good positive correlation)
    
# From the data above one can conclude that if one invests marketing on TV products the Sales will increase for sure.
```
